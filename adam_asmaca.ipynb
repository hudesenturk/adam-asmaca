```python
#Adam asmaca
import random as rnd

kelimeler=[]

hayvan=["kedi","köpek","kuş","goril","tavuk","kaplan"]
esya=["bardak","telefon","klavye","şişe","bardak","fırın","televizyon"]
tasit=["motosiklet","otomobil","otobüs","uçak","vapur","tren","kaykay","metro"]

soru=[hayvan,esya,tasit]

def kelime_sec(indeks):
    kelime=rnd.choice(soru[indeks])
    soru[indeks].remove(kelime)
    return kelime



def cizgi(kelime):
    LCizgi=["-" for i in kelime]

    return LCizgi,kelime


def harf_tahmin(kelime,cizgi,harf):
    for i in range(len(kelime)):
        if harf == kelime[i]:
            cizgi[i]=harf

    return cizgi
print("Kategoriler: 1- Hayvanlar, 2- Eşyalar, 3- Taşıtla")

secim=int(input("Seçiminiz"))-1


bildin=False
while True:
    kelime=kelime_sec(secim)

    line,word=cizgi(kelime)

    for i in line:
        print(i,end=" ")

    for i in range(7):
        harf=input("Harf giriniz: ")


        sonuc=harf_tahmin(word,line,harf)

        print(sonuc)

        if "-" not in line:
            bildin=True
            break


    if bildin==True:
        print("Tebrikler, kazandın")
        break
    else:
        print("Kaybettiniz...")
        break
```
